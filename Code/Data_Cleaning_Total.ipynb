Clean CapIQ.

Goal: Extract all mergers and acquisitions of public companies by private investment companies identified as “going private,” or “leveraged buyout” between January 1980 and October 2023.

Output: List of CUSIPS of Target

```python
# Sift for particular rows that satisfy the below conditions:
# - Buyer is a Private Investment Firm
# - Target is a Public Company
# - Transaction is either Closed or Cancelled
# - Transaction type is a Merger/Acquisition
CapIQ = pd.read_csv('/content/drive/MyDrive/Senior Fall/Senior Thesis/Data/data_capIQ_pub2priv.csv')
CapIQ = CapIQ.loc[CapIQ['Company Type [Buyers/Investors]'].str.contains("Private Investment Firm")]
CapIQ = CapIQ.loc[CapIQ['Company Type [Target/Issuer]'].str.contains("Public Company")]
CapIQ = CapIQ.loc[CapIQ['Transaction Status'] != 'Announced']
CapIQ = CapIQ.loc[CapIQ['Transaction Types'] == 'Merger/Acquisition']

# Drop the columns of which we have no use
CapIQ = CapIQ.drop('Company Type [Buyers/Investors]', axis=1)
CapIQ = CapIQ.drop('Company Type [Target/Issuer]', axis=1)
CapIQ = CapIQ.drop('Transaction Types', axis=1)

# Obtain GVKEY for all Target companies and output a text file
# The outputted text file is used to obtain data from Compustat
Target_GVKEY = CapIQ['Compustat GVKEY [Target/Issuer]'].tolist()
Target_GVKEY = [x for x in Target_GVKEY if x != '-']

with open("data_Target_GVKEY.txt", 'w') as file:
        for row in Target_GVKEY:
            s = "".join(map(str, row))
            file.write(s+'\n')

# Obtain Tickers for Target companies
def substring_after(s, delim):
    return s.partition(delim)[2]

CapIQ['Ticker'] = [substring_after(x, ":") for x in CapIQ['Exchange:Ticker']]
```

```python
SDC = pd.read_csv('/content/drive/MyDrive/Senior Fall/Senior Thesis/Data/data_SDC_pub2priv.csv')

# Sift for particular rows that satisfy the below conditions:
# - Acquiror is a Private Finanical Sponsor / Leveraged Buyout Firm
# - Target is a Public Company
# - Acquiror and Target are both in the United States -- domestic M&A
# - Transaction is either Closed or Cancelled
# - Transaction type is a Merger/Acquisition
SDC = SDC.loc[SDC['Deal Status'] == 'Completed']
SDC = SDC.loc[SDC['Acquiror is a Financial Sponsor (Yes/No Flag)'] == 'Yes']
SDC = SDC.loc[SDC['Acquiror is a Leverage Buyout Firm'] == 'Yes']
SDC = SDC.loc[SDC['Acquiror Public Status'] == 'Priv.']
SDC = SDC.loc[SDC['Target Public Status'] == 'Public']
SDC = SDC.loc[SDC['Acquiror Nation'] == 'United States']
SDC = SDC.loc[SDC['Target Nation'] == 'United States']

# Drop the columns of which we have no use
SDC = SDC.drop('Deal Status', axis=1)
SDC = SDC.drop('Acquiror is a Financial Sponsor (Yes/No Flag)', axis=1)
SDC = SDC.drop('Acquiror is a Leverage Buyout Firm', axis=1)
SDC = SDC.drop('Acquiror Public Status', axis=1)
SDC = SDC.drop('Target Public Status', axis=1)
SDC = SDC.drop('Acquiror Nation', axis=1)
SDC = SDC.drop('Target Nation', axis=1)

# Rename column(s)
SDC.rename(columns={'Target Ticker Symbol': 'Ticker'}, inplace=True)

# Obtain CUSIPs for all Target companies and output a text file
# The outputted text file is used to obtain data from Compustat
Target_CUSIP = SDC['Target CUSIP'].tolist()

with open("data_Target_CUSIP.txt", 'w') as file:
        for row in Target_CUSIP:
            s = "".join(map(str, row))
            file.write(s+'\n')
```

```python
SDC['Date Announced']= pd.to_datetime(SDC['Date Announced'])
CapIQ['All Transactions Announced Date'] = pd.to_datetime(CapIQ['All Transactions Announced Date'])

SDC['Announced Year'] = SDC['Date Announced'].dt.year
CapIQ['Announced Year'] = CapIQ['All Transactions Announced Date'].dt.year
```

#Create Summary Statistics of the Public-to-Private Sample

### Find unique observations from both the SDC and CapIQ dataframes

```python
SDCShort = SDC[['Announced Year', 'Ticker']]
SDCShort = SDCShort[SDCShort['Ticker'].notna()]
SDCShort = SDCShort.sort_values(by=['Ticker', 'Announced Year'], ascending=[True, True])
SDCShort = SDCShort.drop_duplicates(subset='Ticker', keep='last')
SDCShort = SDCShort.drop(SDCShort.index[0])

CapShort = CapIQ[['Announced Year', 'Ticker']]
CapShort = CapShort[CapShort['Ticker'].notna()]
CapShort = CapShort.sort_values(by=['Ticker', 'Announced Year'], ascending=[True, True])
CapShort = CapShort.drop_duplicates(subset='Ticker', keep='last')
CapShort = CapShort.drop(CapShort.index[0])

Unique = SDCShort.merge(CapShort, on='Ticker', how='outer')
Unique = Unique[Unique['Ticker'].notna()]
Unique['Announced Year'] = Unique['Announced Year_x'].combine_first(Unique['Announced Year_y'])
Unique = Unique.drop(columns=['Announced Year_x', 'Announced Year_y'])
Unique = Unique.sort_values(by=['Ticker', 'Announced Year'], ascending=[True, True])
Unique = Unique.drop_duplicates(subset='Ticker', keep='last')
Unique = Unique.drop(Unique.index[0])
```

### Create the Table

```python
# Group the dataframes and calculate the sizes
SDC_grouped = SDCShort.groupby(['Announced Year']).size().reset_index(name='SDC')
CapIQ_grouped = CapShort.groupby(['Announced Year']).size().reset_index(name='CapIQ')
Unique_grouped = Unique.groupby(['Announced Year']).size().reset_index(name='Unique')

result = pd.merge(CapIQ_grouped, SDC_grouped, on='Announced Year', how='outer').fillna(0)
result = pd.merge(result, Unique_grouped, on='Announced Year', how='outer').fillna(0)
result.set_index('Announced Year', inplace=True)
result = result.iloc[1:-1]

# Add a row at the very bottom with the sums
result.loc['Total'] = result.sum(numeric_only=True)
result = result.astype(int)

print(result)

result.to_excel('uniquePub2Priv.xlsx')
```

```python
CF = pd.read_excel('/content/drive/MyDrive/Senior Fall/Senior Thesis/Data/data_cashflow.xlsx')
```

```python
CF = CF.drop('FUND MANAGER', axis=1)
CF = CF.drop('NAME', axis=1)
CF = CF.drop('GEOGRAPHIC FOCUS', axis=1)
CF = CF.drop('INDUSTRIES', axis=1)
CF = CF.drop('FUND ID', axis=1)
CF = CF.drop('FIRM ID', axis=1)
CF = CF.drop('VINTAGE / INCEPTION YEAR', axis=1)
CF.head()
```

```python
CF['TRANSACTION DATE'] = pd.to_datetime(CF['TRANSACTION DATE'])
CF['TRANSACTION YEAR'] = CF['TRANSACTION DATE'].dt.year
```

```python
import matplotlib.pyplot as plt

desired_transaction_types = ['Capital Call', 'Distribution']
CF_filtered = CF[CF['TRANSACTION TYPE'].isin(desired_transaction_types)]
CF_annual = CF_filtered.groupby(['TRANSACTION YEAR', 'TRANSACTION TYPE'])['TRANSACTION AMOUNT'].sum().unstack(fill_value=0)
ax = CF_annual.plot(y=['Capital Call', 'Distribution'], kind='bar', color=['red', 'green'], figsize=(12, 6), width=1)
ax.legend(loc='upper left')
ax.set_title('Preqin Cash Flows 1980 - 2023')
ax.set_ylabel('Amount (USD)')
plt.show()
```

```python
# Group by 'YEAR,' 'QUARTER,' and 'TRANSACTION TYPE,' and sum the 'TRANSACTION AMOUNT'
CF['QUARTER'] = CF['TRANSACTION DATE'].dt.quarter
CF_filtered = CF[CF['TRANSACTION TYPE'].isin(desired_transaction_types)]
CF_quarter = CF_filtered.groupby(['TRANSACTION YEAR', 'QUARTER', 'TRANSACTION TYPE'])['TRANSACTION AMOUNT'].sum().unstack(fill_value=0)
ax = CF_quarter.plot(y=['Capital Call', 'Distribution'], kind='bar', color=['red', 'green'], figsize=(12, 6), width=1)
ax.legend(loc='upper left')
ax.set_title('Preqin Cash Flows 1980 - 2023')
ax.set_ylabel('Amount (USD)')
#x_ticks = [str(year) for year in CF_annual.index.to_timestamp().year]
#ax.set_xticklabels(x_ticks)

plt.show()
```

```python
FV = pd.read_excel('/content/drive/MyDrive/Senior Fall/Senior Thesis/Data/data_fund_values.xlsx')
FV = FV.drop('FUND ID', axis=1)
FV = FV.drop('FIRM ID', axis=1)
FV = FV.drop('FUND SERIES ID', axis=1)
FV = FV.drop('NAME', axis=1)
FV = FV.drop('COUNTRY', axis=1)
FV = FV.drop('GEOGRAPHIC FOCUS', axis=1)
FV = FV.drop('FUND MANAGER', axis=1)
FV['FINAL CLOSE SIZE (USD BN)'] = FV['FINAL CLOSE SIZE (USD MN)'] / 1000
FV = FV.drop('FINAL CLOSE SIZE (USD MN)', axis=1)
FV = FV.groupby('VINTAGE / INCEPTION YEAR')['FINAL CLOSE SIZE (USD BN)'].sum().reset_index()
FV = FV.iloc[:-1]

plt.figure(figsize=(10, 6))
plt.plot(FV['VINTAGE / INCEPTION YEAR'], FV['FINAL CLOSE SIZE (USD BN)'], linestyle='-.', color='green')
plt.ylabel('$ BILLIONS')
plt.title('Preqin Yearly Buyout Fund Values')
plt.grid(False)

FV['CUMULATIVE SIZE (USD BN)'] = FV['FINAL CLOSE SIZE (USD BN)'].cumsum()
plt.figure(figsize=(10, 6))
plt.plot(FV['VINTAGE / INCEPTION YEAR'], FV['CUMULATIVE SIZE (USD BN)'], linestyle='-.', color='green')
plt.ylabel('$ BILLIONS')
plt.title('Preqin Aggregate Buyout Fund Values')
plt.grid(False)
```
