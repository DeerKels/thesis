**Transaction Screening on SDC Platinum**

Date Announced: [1/1/1981 - 10/28/2023]

Target Public Status: Public

Target Nation: United States

Acquiror Public Status: Private

Acquiror is a Leverage Buyout Firm

```python
SDC = pd.read_excel('/content/drive/MyDrive/Senior Fall/Senior Thesis/Data/Separately/SDC.xlsx')

### We only care about rows where 'Deal Status' is 'Completed'
SDC = SDC[SDC['Deal Status'].isin(['Completed'])]

### Drop Rows where the Target has no 'Ticker'
SDC = SDC.dropna(subset=['Target Ticker Symbol'])
Ticker = SDC['Target Ticker Symbol'].tolist()
with open("Ticker.txt", 'w') as file:
        for row in Ticker:
            s = "".join(map(str, row))
            file.write(s+'\n')
```

```python
### Sort by 'Date Announced' in decreasing order
SDC['Date Announced'] = pd.to_datetime(SDC['Date Announced'])
SDC = SDC.sort_values(by='Date Announced', ascending=False)
SDC.head(20)
SDC
```

**Pretransaction Data for SDC Targets**

```python
SDC_pre = pd.read_excel('/content/drive/MyDrive/Senior Fall/Senior Thesis/Data/Separately/SDC_pretransaction.xlsx', converters={'Global Company Key':str})
```

```python
def return_df_from_ticker(ticker, pre):
  return pre.loc[pre['Ticker Symbol'] == ticker]

def get_row_from_ticker(ticker, date_announced, filtered_df):
  return filtered_df.loc[filtered_df['Data Date'] <= date_announced].tail(1)

def return_df_from_gvkey(gvkey, pre):
  return pre.loc[pre['Global Company Key'] == gvkey]

def get_row_from_gvkey(gvkey, date_announced, filtered_df):
    return filtered_df.loc[filtered_df['Data Date'] <= date_announced].tail(1)
```

```python
SDC_filtered = SDC[['Date Announced', 'Target Ticker Symbol']]
Pre_filtered = pd.DataFrame(columns=SDC_pre.columns)

for index, row in SDC_filtered.iterrows():
    date_announced = row['Date Announced']
    ticker = row['Target Ticker Symbol']
    row = get_row_from_ticker(ticker, date_announced, return_df_from_ticker(ticker, SDC_pre))
    Pre_filtered = Pre_filtered.append(row, ignore_index=True)
```

# Let's get the pretty table

```python
SDC_filtered = SDC[['Date Announced', 'Target Ticker Symbol']]
Pre_filtered_SDC = pd.DataFrame(columns=SDC_pre.columns)

for index, row in SDC_filtered.iterrows():
    date_announced = row['Date Announced']
    ticker = row['Target Ticker Symbol']
    row = get_row_from_ticker(ticker, date_announced, return_df_from_ticker(ticker, SDC_pre))
    Pre_filtered_SDC = Pre_filtered_SDC.append(row, ignore_index=True)
```

```python
Pre_filtered_SDC.shape
```

```python
Pre_filtered_SDC = Pre_filtered_SDC.groupby('Fiscal Year').size().reset_index(name='SDC Platinum')
```

```python
cap_pre['Data Date'] = pd.to_datetime(cap_pre['Data Date'])
cap_filtered = CapIQ[['All Transactions Announced Date', 'Compustat GVKEY [Target/Issuer]']]
Pre_filtered_Cap = pd.DataFrame(columns=cap_pre.columns)

for index, row in cap_filtered.iterrows():
    date_announced = row['All Transactions Announced Date']
    gvkey = row['Compustat GVKEY [Target/Issuer]']
    row = get_row_from_gvkey(gvkey, date_announced, return_df_from_gvkey(gvkey, cap_pre))
    Pre_filtered_Cap = Pre_filtered_Cap.append(row, ignore_index=True)
```

```python
Pre_filtered_Cap = Pre_filtered_Cap.groupby('Fiscal Year').size().reset_index(name='Capital IQ')
```

```python
Pre_filtered = Pre_filtered.groupby('Fiscal Year').size().reset_index(name='Unique')
```

```python
Merged = pd.merge(Pre_filtered_Cap, Pre_filtered_SDC, on='Fiscal Year', how='outer')
Merged = pd.merge(Merged, Pre_filtered, on='Fiscal Year', how='outer')
Merged['Capital IQ'].fillna(0, inplace=True)
Merged['SDC Platinum'].fillna(0, inplace=True)
Merged['Capital IQ'] = Merged['Capital IQ'].astype(int)
Merged['SDC Platinum'] = Merged['SDC Platinum'].astype(int)
Merged = Merged.sort_values(by='Fiscal Year')
sum_row = Merged.sum(numeric_only=True)  # This sums only the numeric columns
Merged = Merged.append(sum_row, ignore_index=True)
Merged.iloc[-1, 0] = 'Total'
Merged.set_index('Fiscal Year', inplace=True)
Merged.to_excel('Cap_SDC_Unique.xlsx')
```
