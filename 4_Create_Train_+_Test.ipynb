````python
# Drop columns that are not in Test.columns
selected = selected[Test_X.columns]

# Drop rows w/ more than 30% NAs
selected = selected.dropna(thresh=0.7*len(selected.columns))

# Clean the NAs by taking the average value of that column up to that time
selected['Data Date'] = pd.to_datetime(selected['Data Date'])
selected = selected.sort_values(by='Data Date')
selected = selected.drop(columns=['Ticker Symbol', 'CUSIP', 'Company Name', 'CIK Number', 'Active/Inactive Status Marker', 'Global Company Key', 'Data Date'])

for col in range(len(selected.columns)):
  only_nas_so_far = True
  for row in range(len(selected)):
    if not pd.isna(selected.iloc[row, col]) and only_nas_so_far:
      only_nas_so_far = False

    if pd.isna(selected.iloc[row, col]) and not only_nas_so_far:
      mean_up_to_i = selected[selected.columns[col]].iloc[:row].mean(skipna=True)
      selected.iloc[row, col] = mean_up_to_i

for column in selected.columns:
  selected[column].fillna(value=selected[column].mean(skipna=True), inplace=True)
````

````python
# Drop columns that are not in Test.columns
not_selected = not_selected[Test_X.columns]

# Drop rows w/ more than 30% NAs
not_selected = not_selected.dropna(thresh=0.7*len(not_selected.columns))

# Clean the NAs by taking the average value of that column up to that time
not_selected['Data Date'] = pd.to_datetime(not_selected['Data Date'])
not_selected = not_selected.sort_values(by='Data Date')
not_selected = not_selected.drop(columns=['Ticker Symbol', 'CUSIP', 'Company Name', 'CIK Number', 'Active/Inactive Status Marker', 'Global Company Key', 'Data Date'])
not_selected = not_selected.apply(pd.to_numeric, errors='coerce')

for col in range(len(not_selected.columns)):
  only_nas_so_far = True
  for row in range(len(not_selected)):
    if not pd.isna(not_selected.iloc[row, col]) and only_nas_so_far:
      only_nas_so_far = False

    if pd.isna(not_selected.iloc[row, col]) and not only_nas_so_far:
      mean_up_to_i = not_selected[not_selected.columns[col]].iloc[:row].mean(skipna=True)
      not_selected.iloc[row, col] = mean_up_to_i

for column in not_selected.columns:
  not_selected[column].fillna(value=not_selected[column].mean(skipna=True), inplace=True)

not_selected
````
